# Checking Outliers

```python
# visualize the outliers using box plot
x.boxplot(figsize=[20,8]) 
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```python
# to remove outliers
from scipy.stats import zscore
z=np.abs(zscore(x))
z.shape

threshold=3
print(np.where(z>3))

x_new=x[(z<3).all(axis=1)]
print("total_dropped_rows",x.shape[0]-x_new.shape[0])

y=y[(z<3).all(axis=1)]
y.shape
```

# Training process begins

Finding best random state

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

Best random state

```python
maxAccu=0 #maximum accuracy
maxRS=0 #best random state value for which max accuracy is acheived 

for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=i)
    LR = LogisticRegression()
    LR.fit(x_train,y_train)    #fitting the data will train your model
    predrf=LR.predict(x_test)  #this is the predicted target variable
    acc=accuracy_score(y_test,predrf) #accuracy score 
    print("accuracy: ",acc, "random_state: ",i)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
        print("accuracy: ",maxAccu, "random_state: ",i)

```

```python
print("Best accuracy is", maxAccu,"on Random State", maxRS)
```

We have found bestrandom_state value as 173. We will create our train_test_split using this random_state

# Creating train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=173)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Logistic Regression

```python
LR = LogisticRegression()
LR.fit(x_train,y_train)
y_pred=LR.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred)*100)  #testing accuracy
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_dt)*100) #testing accuracy
print(confusion_matrix(y_test,y_pred_dt))
print(classification_report(y_test,y_pred_dt))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_rf)*100) #testing accuracy
print(confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

# SVC

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)

y_pred_svc = svc.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_svc)*100) #testing accuracy
print(confusion_matrix(y_test,y_pred_svc))
print(classification_report(y_test,y_pred_svc))
```

```python
# Plot the ROC AUC curve
from sklearn.metrics import plot_roc_curve
disp = plot_roc_curve(dt,x_test,y_test)
plot_roc_curve(LR,x_test,y_test,ax=disp.ax_)
plot_roc_curve(svc,x_test,y_test,ax=disp.ax_)
plot_roc_curve(rf,x_test,y_test,ax=disp.ax_)
plt.show()
```

```python
#Logistic Regression, Random Forest and SVC are producing good accuracy. 
#Now we will check cross validation score as well for overfitting. 
```

# Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV

#Creating parameter list to pass in GridsearchCV
param_grid = { 'C': np.arange(0.001,10,0.1),
              'gamma': np.arange(0.1,1,0.1)
             }
```

```python
gridsearch = GridSearchCV(svc, param_grid)
gridsearch.fit(x_train, y_train) #fitting the data in model
gridsearch.best_params_ #printing the best parameter found by GridSearchCV
```

```python
svc = SVC(C=2.5, gamma=0.2)
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
cfm = confusion_matrix(y_test,y_pred)
print(classification_report(y_test,y_pred))
```

```python
GCV = GridSearchCV(SVC(),param_grid,cv=5,scoring='accuracy')
GCV.fit(x_train,y_train)
GCV.best_params_
GCV_pred=GCV.best_estimator_.predict(x_test) #predicting with best parameters
accuracy_score(y_test,GCV_pred) #checking final accuracy
```

# ROC AUC plot

```python
from sklearn.metrics import plot_roc_curve
plot_roc_curve(GCV.best_estimator_,x_test,y_test)
plt.title("ROC AUC PLOT")
plt.show()
```
